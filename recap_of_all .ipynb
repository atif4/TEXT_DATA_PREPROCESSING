```python
get_duplicates_dic = {}
def get_and_remove_duplicates(list_c_ids):
    for list_c_id in list_c_ids:
        dup_sum = df[list_c_id].duplicated().sum()
        get_duplicates_dic[list_c_id] = dup_sum
        print(list_c_id,'has',dup_sum,'values.')
        #drop_duplocates = df[list_c_id].drop_duplicates() # all duplicates values has been removed
        #print(b,drop_duplocates)
```

```python
#Compute the sorted, common elements in x and y
np.intersect1d(reindex_df['employee_residence'],reindex_df['company_location'])
```

```python
#Compute a boolean array indicating whether each element of x is contained in y
np.in1d(reindex_df['employee_residence'],reindex_df['company_location'])
```

```python
#Set difference, elements in x that are not in y
np.setdiff1d(reindex_df['employee_residence'],reindex_df['company_location'])
```
